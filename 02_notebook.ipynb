```python
import matplotlib
import matplotlib.pyplot as plt

import numpy as np

%config InlineBackend.figure_format='retina'
%matplotlib inline

matplotlib.rc('figure', figsize=(12,9))
matplotlib.rc('axes', edgecolor='k')
matplotlib.rc('grid', color='k', linewidth=0.5, linestyle=':')

matplotlib.rc('image', aspect='auto', interpolation='nearest', cmap='inferno')
```

```python
import scipy.linalg as la
```

```python
N = 1024
```

```python
A = np.random.standard_normal((N, N))
A2 = np.dot(A, A.T)

_, evecs = la.eigh(A2)
```

```python
i = np.arange(N)
evals1 = np.exp(-0.5 * (i / 500.0)**2)
evals2 = np.exp(-0.5 * (i / 50.0)**2)
```

```python
M1 = np.dot(evecs, (evals1[:, np.newaxis] * evecs.T))
M2 = np.dot(evecs, (evals2[:, np.newaxis] * evecs.T))
```

```python
evalsp1 = la.eigvalsh(M1)[::-1]
evalsp2 = la.eigvalsh(M2)[::-1]
```

```python
plt.plot(evalsp1, label="Calculated")
plt.plot(evals1, label="Exact")
plt.legend()
plt.semilogy()
plt.xlabel("Eigenmode")
plt.ylabel(r"$\lambda$")

#plt.ylim(1e-16, 10)
```

```python
plt.plot(evals2, color="C0", label="Exact")
plt.plot(evalsp2, color="C1", label="Calculated")
plt.plot(-evalsp2, color="C1", ls="--")
plt.axhline(1e-16, color="k", ls=":")
plt.legend()
plt.semilogy()
plt.xlabel("Eigenmode")
plt.ylabel(r"$\lambda$")


plt.semilogy()
```
